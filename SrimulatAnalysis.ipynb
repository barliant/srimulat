```python
%pylab
%matplotlib inline
import pandas as pd
from pandas import Series
# Make the graphs a bit prettier, and bigger
pd.set_option('display.mpl_style', 'default') 
pd.set_option('display.width', 5000) 
pd.set_option('display.max_columns', 60)
figsize(15,5)
diskusi = pd.read_csv("/tmp/cleanFile.txt",delimiter="\t",parse_dates='date',index_col='date')
diskusi['sender'].value_counts().plot(kind="bar")
```

```python
# Waktu Posting
import datetime
import matplotlib.pyplot as plt

dini_hari = diskusi["sender"][datetime.time(00,00,00): datetime.time(06,00,00)].count()
pagi = diskusi["sender"][datetime.time(06,01,00): datetime.time(11,00,00)].count()
siang = diskusi["sender"][datetime.time(11,01,00): datetime.time(15,00,00)].count()
sore = diskusi["sender"][datetime.time(15,01,00): datetime.time(18,00,00)].count()
malam = diskusi["sender"][datetime.time(18,01,00): datetime.time(23,00,00)].count()

labels = 'Dini Hari', 'Pagi', 'Siang', 'Sore', 'Malam'
sizes = [dini_hari, pagi, siang, sore, malam]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral', 'grey']
explode = (0, 0, 0, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=False)
# Set aspect ratio to be equal so that pie is drawn as a circle.
plt.axis('equal')



print "Dini Hari\t" + str(dini_hari)
print "Pagi\t\t" + str(pagi_hari)
print "Siang\t\t" + str(siang)
print "Sore\t\t" + str(sore)
print "Malam\t\t" + str(malam)

 
```
